```python
import pandas as pd
from apyori import apriori
```

```python
import numpy as np
```

```python
alcohol_transactions = pd.read_csv('/content/sample_data_apriori.csv')
alcohol_transactions.head()
```

```python
alcohol_transactions.drop('transaction_id',axis =1, inplace=True)
```

```python
# modify data for apyori algorithm
transactions_list =[]
for index, row in alcohol_transactions.iterrows():
  transactions_list.append(list(row.dropna()))
```

```python
# Apply Apriori Algorithm

apriori_rules = apriori(transactions_list,
                        min_support = 0.003,
                        min_confidence = 0.2,
                        min_lift = 3,
                        min_length = 2,
                        max_length = 2)
```

```python
apriori_rules = list(apriori_rules)
```

```python
apriori_rules[0]
```

```python
# Convert output to DataFrame

product_1 = [list(rule[2][0][0])[0] for rule in apriori_rules]
product_2 = [list(rule[2][0][1])[0] for rule in apriori_rules]
confidence = [rule[2][0][2] for rule in apriori_rules]
lift = [rule[2][0][3] for rule in apriori_rules]

apriori_df  = pd.DataFrame({'product_1':product_1,
                            'product_2':product_2,
                            'confidence':confidence,
                            'lift':lift})
apriori_df.sort_values(by='lift', ascending=False)
```
